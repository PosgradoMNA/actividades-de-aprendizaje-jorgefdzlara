## **Regresión Lineal Simple y Regresión Lineal Múltiple**

La regresión lineal simple, se refiere a una variable independiente para realizar una predicción.

La regresión lineal múltiple, se refiere a múltiples variables independientes para realizar una predicción.

La **Regresión Lineal Simple (SLR)** es un método que nos ayuda a conocer la relación entre dos variables:
1. El predictor (independiente) variable x 
2. El objetivo (dependiente) variable y.

**Gráfico de regresión**

Cuando hablamos de regresión lineal simple, una forma excelente para visualizar el ajuste de nuestros modelos es el utilizar gráficos de regresión.

Éste gráfico va a mostrar una combinación de puntos de datos dispersos (un gráfico de dispersión), así como la línea de regresión lineal ajustada que atraviesa los datos. Esto nos dará un estimado razonable de la relación entre las dos variables, la fuerza de la correlación, así como la dirección (correlación positiva o negativa).

Vamos a visualizar  highway-mpg  como un predictor potencial de la variable precio:

```python
width = 12
height = 10
plt.figure(figsize=(width, height))
sns.regplot(x="highway-mpg", y="price", data=df)
plt.ylim(0,)
```

Podemos observar de esta gráfico que el precio es correlacionado de forma negatica a  highway-mpg ya que la pendiente de regresión es negativa.

Una cosa a tener en mente que cuando se observa un gráfico de regresión es prestar atención a qué tan dispersos están los puntos de datos alrededor de la línea de regresión. Esto va a darte un buen indicador de la varianza de los datos y si el modelo lineal será el mejor ajuste o no. Si los datos están muy alejados de la línea, este modelo lineal, pudiera no ser el mejor para estos datos.

Vamos a comparar este gráfico con el gráfico de regresión de "peak-rpm".

Comparando el gráfico de regresión de "peak-rpm" y  "highway-mpg" vemos que los puntos para "highway-mpg" son más cercanos a la línea generada, en promedio, decrementa. Los puntos para "peak-rpm" tienen más dispersión alrededor de la línea predecida y es más difícil determinar si los puntos están aumentando o decrementando como "peak-rpm" incrementa.

**Gráfico Residual**

Una buena forma de visualizar la varianza de los datos es utilizar el gráfico residual.

¿Qué es residual?

La diferencia entre el valor observado (y) y el valor predecido (Yhat) es llamado el residual (e). Cuando observamos a la gráfica de regresión, el residual es la distancia desde el punto de datos a la línea de regresión ajustada.

Entonces, ¿Qué es la gráfica residual?

Una gráfica residual es la gráfica que muestra los residuales en el eje y vertical y la variable independiente en el eje x horizontal.

¿A qué le prestamos atención cuando observamos el gráfico residual?

Vamos a observar la dispresión de los residuos:

* Si los puntos en un gráfico de residuos están dispersos de forma aleatoria alrededor del eje x, entonces el modelo lineal es apropiado para los datos.

¿Por qué es esto?

Residuales dispersos de forma aleatoria, significan que la varianza es constante, y por lo tanto el modelo lineal es un buen ajuste para estos datos.


```python
width = 12
height = 10
plt.figure(figsize=(width, height))
sns.residplot(x=df['highway-mpg'],y=df['price'])
plt.show()
```

¿Qué nos dice este gráfico?

Podemos ser de este gráfico residual que los residuos no están dospersos aleatoriamente alrededor del eje x, llevándonos a creer que tal vez un modelo no lineal es más apropiado para estos datos.

## **Regresión Polinomial y Pipelines**

**Regresión polinomial es un caso particular del modelo de regresión lineal general o modelos de regresión lineal múltiples.

Obtenemos relaciones no lineales al elevar al cuadrado o definiendo términos de mayor orden en las variables predictoras.

Hay diferentes órdenes de una regresión polinomial:

image.png

Vimos anteriormente que un modelo lineal no proporciona el mejor ajuste mientras utilizamos "highway-mpg" la variable predictor. Entonces vamos a ver si podemos probar ajustar el modelo polinomial a los datos.

Vamos a utilizar la siguiente función para graficar los datos:

```python
def PlotPolly(model, independent_variable, dependent_variabble, Name):
    x_new = np.linspace(15, 55, 100)
    y_new = model(x_new)

    plt.plot(independent_variable, dependent_variabble, '.', x_new, y_new, '-')
    plt.title('Polynomial Fit with Matplotlib for Price ~ Length')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Price of Cars')

    plt.show()
    plt.close()
```

Vamos a obtener las variables:

```python
x = df['highway-mpg']
y = df['price']
```

Vamos a ajustar el polinomio utilizando la función  **polyfit**, luego utilizaremos la función **poly1d** para presentar la función polinomial.

```python
f = np.polyfit(x, y, 3)
p = np.poly1d(f)
print(p)
```

Vamos a graficar la función:

```python
PlotPolly(p, x, y, 'highway-mpg')
```

```python
np.polyfit(x, y, 3)
```

Ya podemos observar de la gráfica que este modelo polinomial se comporta mejor que el modleo lineal. Esto es porque la función polinomial generada, acerta más de los puntos de datos.

```python
f1 = np.polyfit(x, y, 11)
p1 = np.poly1d(f1)
print(p1)
PlotPolly(p1,x,y, 'Highway MPG')
```

# **Pipeline**

Los Pipelines de datos simplifican los pasos en el procesamiento de datos, Utilizamos el módulo de **Pipeline** para crear un pipeline. También usamos **StandardScaler** como un paso en nuestro pipeline.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

Creamos un pipeline con una lista de tuplas, incluyendo el nombre del modelo o estimador y su constructor correspondiente.

```python
Input=[('scale',StandardScaler()), ('polynomial', PolynomialFeatures(include_bias=False)), ('model',LinearRegression())]
```

Ingresamos la lista como un argumento al constructor del pipeline:

```python
pipe=Pipeline(Input)
pipe
```

Primero convertimos el tipo de datos Z a tipo flotante y así evitar alertas de conversión que aparecen como resultado de StandardScaler tomando entradas flotantes.

Después, podemos normalizar los datos, realizar una transformación y ajustar el modelo simultaneamente.

```python
Z = Z.astype(float)
pipe.fit(Z,y)
```

De forma similar, podemos normalizar los datos, realizar una transformación y producir una predicción simultáneamente.

```python
ypipe=pipe.predict(Z)
ypipe[0:4]
```

# **4. Medidas para evaluación en muestras**

Cuando evaluamos nuestros modelos, no solo queremos visualziar los resultados, pero también queremos una medida cuantitativa para determinar qué tan preciso es nuestro modelo.

Dos medidas muy importantes que son utilizadas frecuentemente en Estadística para determinar la precisión de un modelo son:

* **R^2/R-squared**
* **Mean Squared Error (MSE)**

**R- Cuadrada**

R cuadrada, conocido también como el coeficiente de determinación, es una medida que indica qué tan cerca de los datos se encuentra la línea de regresión ajustada.

El valor de la R-cuadrada es el porcentaje de la variación de la variable de respuesta (y) que es explicada por un modelo lineal.

**Mean Squared Error (MSE)**

El error cuadrado promedio, mide el promedio del cuadrado de los errores. Que es la diferencia entre el valor actual (y) y el valor estimado (ŷ)

**Modelo 1: Regresión Lineal Simple**

Vamos a calcular R^2:


```python

#highway_mpg_fit
lm.fit(X, Y)
# Find the R^2
print('The R-square is: ', lm.score(X, Y))
```

Podemos decir que ~49.659% de la variación del precio es explicada por este modelo lineal simple "horsepower_fit".

Vamos a calcular el MSE:

Podemos predecir la salida utilizando el modelo predecido, donde X es la variable de entrada.

```python
Yhat=lm.predict(X)
print('La salida de nuestro primer valor predicho es: ', Yhat[0:4])
```

Vamos a importar la función **mean_squared_error** del módulo de **metrics**:

```python
from sklearn.metrics import mean_squared_error
```

Podemos comparar los resultados predichos como los resultados actuales:

```python
mse = mean_squared_error(df['price'], Yhat)
print('El error promedio de los cuadrados del precio y valores predichos es: ', mse)
```
